# Flags

g - todas as ocorrências

i - case insensitive

m - opera em múltiplas linhas. Atua em conjunto com lida com  os caracteres de inicio e fim (^ e $)

Pipe | - indica mais de uma opção (OR)


# Metacaracteres e Quantificadores
^ - conjunto negado 

. - qualquer caractere

? - zero ou uma ocorrência

* - zero ou mais ocorrência

+ - uma ou mais ocorrências

{n, m} - de n até m ocorrências

Para especificar a forma literal do metacaractere/quantificador use o escape \. Ex. \.

# Âncoras (delimitam a ação da nossa regex)

\b - indica que as extremidades da regex nao possuem espaço ou outras palavras.

^ -  início de linha (exige flag m habilitada)

$ - fim de linha (exige flag m habilitada)

# Conjuntos []

Múltiplos valores que podem dar match. Ex. Ex. [a-z]. Negação de um conjunto de valores é feita com ^. Ex. [^A-Za-z0-9]

# Grupos () 

Regras isoladas que podem ser reutilizadas e/ou referenciadas. Ex. Na regex 
/\d{2}(\/?)\d{2}?\1\d{4}/g há o grupo (\/?) que é referenciado adiante por \1, dispensando repetição. O número indica a ordem de criação do grupo.

# Grupos especiais (?:) (?=) e (?!)

(?:) - Non-capturing groups. 

(?=) - Positive lookahead. Regex só dá match em palavras que à sua frente possuam esse grupo. Ex. /([a-z]+)(?=,)/gi. Se a string for "Penso, logo existo", o match ocorrerá em "Penso,".

(?!) - Negative lookahead. Regex só dá match em palavras que à sua frente não possuam esse grupo. Ex. /([a-z]+)(?!,)/gi. Se a string for "Penso, logo existo", o match ocorrerá em "logo existo".

(?i): case-insensitive marker. Ex. (?i)TABLE

# Regex no Python

- re.compile: compile a regular expression pattern provided as a string into a regex pattern object

- re.match: tenta encontrar um padrão dentro de uma string. Se não encontrar, retorna None

- re.findall: retornar uma lista de elementos com aquele padrão.

- re.sub: busca um padrão para substituí-lo por outro

- re.finditer: iteração de strings em busca de um dado padrão. Retorna também o índice dessas strings. 

- re.search: busca por um padrão e retorna informações completas, como a posição inicial e a final daquele padrão dentro de uma string,

```python
import re

TEXT = 'Penso, logo existo'
PATTERN_POSITIVE_LOOKAHEAD = r'([a-z]+)(?=,)'
PATTERN_NEGATIVE_LOOKAHEAD = r'([a-z]+)(?!,)'

reg_positive = re.compile(PATTERN_POSITIVE_LOOKAHEAD, flags=re.IGNORECASE)
print(reg_positive.findall(TEXT))

for m in reg_positive.finditer(TEXT):
  print(m.groups())

reg_negative = re.compile(PATTERN_NEGATIVE_LOOKAHEAD, flags=re.IGNORECASE)
print(reg_negative.findall(TEXT))
```

```python
print(re.findall('the', 'The sources informed the police of tHe thieves.', flags=re.IGNORECASE))
print(re.search('the', 'The sources informed the police of tHe thieves.', re.IGNORECASE))
print(re.match('the', 'The sources informed the police of tHe thieves.', re.IGNORECASE))
print(re.match('not found', 'The sources informed the police of tHe thieves.', re.IGNORECASE))
```

```python
print(re.sub(PATTERN_POSITIVE_LOOKAHEAD, 'REPLACED', TEXT))
```

```python
s = '''Regex 
Multiline'''

print(re.findall('^\w+', s))
print(re.findall('^\w+', s, flags=re.MULTILINE))
```
